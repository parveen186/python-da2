```python
df[['Category','Sales']].groupby('Category').sum()
```

```python
#FINDOUT the average profit made by each region in each category
#group by region, category
#profit-average
df[['Category','Region','Profit']].groupby(['Category','Region']).mean()
```

```python
#find out average,minimum,max sales in each region
gt=df[['Region','Sales']].groupby('Region')
gt.mean()
gt.max()
gt.min()
```

```python
gt.aggregate({'Sales':['min','max','mean']})
```

```python
#find out average and total sales and avg and max proft for each category in each region
gr=df[['Region','Category','Sales','Profit']].groupby(['Region','Category'])
```

```python
gr.aggregate({'Sales':['mean','sum'],'Profit':['mean','max']})
```

```python
#find out top2 subcategory based on total profit made in each region
gr=df[['Region','Sub-Category','Profit']].groupby(['Region','Sub-Category'])
```

```python
gr.sum()
```

```python
grnew=gr.sum().sort_values(by=['Region','Profit'],ascending=False)
```

```python
grnew
```

```python
grnew.groupby('Region').head(2)
```

```python
#find out least profitable subcategory in each year 
import datetime as dt
df['Year']=pd.to_datetime(df['Order Date']).dt.year
```

```python
grnew=df[['Year','Sub-Category','Profit']].groupby(['Year','Sub-Category'])
```

```python
grnew.sum().sort_values(by=['Year','Profit'],ascending=False)
```

```python
# pivot_table
# find out of total sales and minimum sale,total profit and average discount for each category in each year
df.pivot_table(index=['Year','Category'],values=['Sales','Profit','Discount'],aggfunc={'Sales':['sum','min'],'Profit':['sum'],'Discount':['mean']})
```
